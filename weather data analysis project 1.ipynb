```python
# Replace column name
df.rename(columns={'Rel Hum_%':'Humidity','Weather':'Weather_Type'},inplace=True)
```

```python
df['Date/Time']=pd.to_datetime(df['Date/Time'])
df['Month']=df['Date/Time'].dt.month
df['hour']=df['Date/Time'].dt.hour
df['Year']=df['Date/Time'].dt.year
df['Day']=df['Date/Time'].dt.day
```

```python
df['Month']=df['Month'].map({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'June',7:'July',8:'Aug',9:'Sept',10:'Oct',11:'Nov',12:'Dec'})
```

```python
# Drop two unnecessary columns
df.drop(['hour','Year'],axis=1,inplace=True)
```

# Data Visualization

```python
# Humidity distribution
fig,ax= plt.subplots(2,2,figsize=(10, 10))
plt.suptitle('Distribution Of various meterological factors',fontsize=15,fontweight='bold')
ax[0,0].set_facecolor('lightblue')
ax[0,1].set_facecolor('lightblue')
sns.histplot(df['Humidity'],color='Red',kde=True,ax=ax[0,0])
ax[0,0].set_title('Humidity Distribution',fontsize=15)

sns.histplot(df['Temp_C'],color='Red',kde=True,ax=ax[0,1])
ax[0,1].set_title('Temperature Distribution',fontsize=15)

sns.violinplot(x=df['Wind Speed_km/h'],color='Red',inner='stick',ax=ax[1,0])
ax[1,0].set_title('windspeed Distribution',fontsize=15)
ax[1,0].set_facecolor('lightblue')

sns.distplot(df['Visibility_km'],color='Red',kde=True,ax=ax[1,1])
ax[1,1].set_title('Visibility distribution',fontsize=15)
ax[1,1].set_facecolor('lightblue')
```

```python
plt.title("Weather Types",fontsize=20,fontweight='bold')
weather_types=df['Weather_Type'].value_counts().head(10)
weather_types.plot(kind='bar',color='black',figsize=(12,6))
plt.xticks(rotation=0)
plt.xticks(fontsize=9)
plt.show()
```

the above plot represent the types of weather follows over times and we found mostly weather is clear or cloudy which
means an ideal level of weather maintains most of the times

```python
plt.figure(figsize=(12,5))
sns.lineplot(x='Month',y='Temp_C',data=df)
plt.title('Temperature over months',fontsize=20,fontweight='bold')
plt.xlabel('')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.histplot(x=df['Press_kPa'],kde=True,color='brown')
plt.title('Distribution of atmospheric pressure',fontsize=20,fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.title('Dew Point Temperature Distribution',fontsize=20,fontweight='bold')
dew_temp_distribution = sns.boxplot(x=df['Dew Point Temp_C'],color='lightgreen')
dew_temp_distribution
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.scatterplot(x=df['Temp_C'],y=df['Dew Point Temp_C'],color='green')
plt.title('Temperature vs Dew Point Temperature',fontweight='bold',fontsize=20)
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x=df['Temp_C'],y=df['Humidity'],color='red')
plt.title('Temperature vs Humidity',fontweight='bold',fontsize=20)
plt.tight_layout()
plt.show()
```

```python
# Temperature over time
plt.figure(figsize=(10,4))
sns.scatterplot(x=df['Date/Time'],y=df['Temp_C'],linestyle='dashdot',color='black')
plt.title('Temperature over time',fontsize=17,fontweight='bold')
plt.show()
```

```python
# heatmap of correlation matrix with annotation
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(),annot=True,linewidths=.4,cmap='jet')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='Day',y='Wind Speed_km/h',data=df)
plt.title('Day vs Temperature',fontsize=17,fontweight='bold')
plt.show()
```
